```python
# Newton’s Law of Cooling Simulation (Multiple Lambda Values)

## Objective
To simulate temperature decay using Newton’s Law of Cooling and analyse how different cooling constants (λ) affect the rate of temperature change.

## Key Insights
- Higher λ values result in faster cooling rates
- Temperature decay follows an exponential trend toward ambient temperature
- Smaller λ values produce slower, more gradual cooling
- Euler method provides a good numerical approximation of temperature evolution

import numpy as np
import matplotlib.pyplot as plt

# Constants
T0 = 50              # Initial temperature (°C)
T_env = 0           # Ambient temperature (°C)
dt = 0.1             # Time step (seconds)
temp_tolerance = 0.1 # Stop when temperature is close enough to ambient
max_time = 1000      # Maximum simulation time (seconds)

# Lambda values to simulate
lambda_values = [0.1, 0.5, 1, 2, 3, 4, 5]

plt.figure(figsize=(10,6))

for k in lambda_values:
    # Initialize temperature and time arrays for each lambda
    T = [T0]
    t = [0]

    # Euler method simulation
    while abs(T[-1] - T_env) > temp_tolerance and t[-1] < max_time:
        dTdt = -k * (T[-1] - T_env)
        T_next = T[-1] + dTdt * dt
        T.append(T_next)
        t.append(t[-1] + dt)

    plt.plot(t, T, label=f'λ = {k}')

# Plot ambient temperature line
plt.axhline(T_env, color='r', linestyle='--', label='Ambient Temperature')

plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title("Newton's Law of Cooling for Various λ (Euler Method)")
plt.legend()
plt.grid(True)
plt.show()
```
