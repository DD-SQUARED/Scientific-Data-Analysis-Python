```python
# Light Intensity vs Polarization Angle Analysis

## Objective
To analyse how light intensity varies with angle for different polarizer configurations and verify polarization laws.

## Key Insights
- Intensity follows expected cosine-squared relationship with angle
- Additional polarizers modify intensity behaviour significantly
- Results align with theoretical predictions of light polarization
- Brewster’s angle behaviour observed in intensity variation


 Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Upload and read the CSV file
from google.colab import files
uploaded = files.upload()

# Load the CSV file into a DataFrame
df_two_polarizers = pd.read_csv('two_polarizers.csv')

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(df_two_polarizers['Angle'], df_two_polarizers['Intensity'], marker='o', linestyle='-', color='b')
plt.title('Intensity vs. Angle for Two Polarizers')
plt.xlabel('Angle (degrees)')
plt.ylabel('Intensity (V)')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Upload and read the CSV file
from google.colab import files
uploaded = files.upload()

# Load the CSV file into a DataFrame
df_three_polarizers = pd.read_csv('three_polarizers.csv')

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(df_three_polarizers['Angle'], df_three_polarizers['Intensity'], marker='o', linestyle='-', color='r')
plt.title('Intensity vs. Angle for Three Polarizers')
plt.xlabel('Angle (degrees)')
plt.ylabel('Intensity (V)')
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Upload and read the CSV file
from google.colab import files
uploaded = files.upload()

# Load the CSV file into a DataFrame
df_brewster_angle = pd.read_csv('brewster_angle.csv')

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(df_brewster_angle['Angle'], df_brewster_angle['Total Light'], marker='o', linestyle='-', color='g', label='Total Light')
plt.plot(df_brewster_angle['Angle'], df_brewster_angle['Polarized Light'], marker='x', linestyle='--', color='m', label='Polarized Light')
plt.title('Light Intensity vs. Angle for Brewster’s Angle')
plt.xlabel('Angle (degrees)')
plt.ylabel('Light Intensity (V)')
plt.legend()
plt.grid(True)
plt.show()
```
